## **Analisando os eventos que ocorreram nesses períodos:**

### **Análise do Aumento do Preço do Petróleo em 1990**
#### **Contexto Histórico**
Em 1990, o mercado global de petróleo sofreu uma grande perturbação devido a eventos geopolíticos significativos na região do Golfo Pérsico. O evento mais marcante foi a **invasão do Kuwait pelo Iraque** em 2 de agosto de 1990, desencadeando o que ficou conhecido como a **Guerra do Golfo**.

#### **Impacto no Mercado de Petróleo**
1. **Interrupção da Produção**:
   - O Kuwait, um dos principais produtores de petróleo da OPEP, teve sua produção completamente interrompida pela invasão.
   - O Iraque também enfrentou sanções econômicas impostas pela comunidade internacional, limitando sua capacidade de exportar petróleo.

2. **Ameaça à Estabilidade Regional**:
   - O Golfo Pérsico é uma das regiões mais estratégicas para a produção e exportação de petróleo.
   - A possibilidade de escalada do conflito e o risco de interrupções em outras rotas e países produtores (como Arábia Saudita) causaram pânico nos mercados.

3. **Aumento nos Preços**:
   - Antes da invasão, o preço médio mensal do petróleo Brent estava em torno de **US17 por barril** (julho de 1990).
   - Em outubro de 1990, o preço atingiu **US41 por barril**, um aumento de mais de 100% em apenas três meses.

#### **Fatores Contribuintes**
1. **Volatilidade nos Estoques e Produção**:
   - A Organização dos Países Exportadores de Petróleo (OPEP) enfrentou dificuldades para coordenar respostas eficazes à crise.
   - Apesar de tentativas de aumentar a produção em outros países membros, as incertezas políticas mantiveram os preços altos.

2. **Sentimento do Mercado**:
   - A guerra trouxe uma incerteza global significativa, afetando tanto países consumidores quanto produtores.
   - A especulação sobre a continuidade do fornecimento de petróleo aumentou a volatilidade dos preços.

### **Análise do Aumento do Preço do Petróleo em 2008**

#### **Contexto Histórico**
Em 2008, o mercado global de petróleo experimentou uma volatilidade sem precedentes, com os preços atingindo níveis recordes no meio do ano e caindo drasticamente nos meses subsequentes. Essas flutuações foram influenciadas por uma combinação de fatores econômicos, geopolíticos e especulativos.

#### **Impacto no Mercado de Petróleo**
1. **Pico dos Preços**:

2. **Queda Abrupta**:

#### **Fatores Contribuintes**
1. **Aumento da Demanda Global**:
   - Países em rápido desenvolvimento, como China e Índia, apresentaram um crescimento econômico acelerado, aumentando significativamente a demanda por petróleo.

2. **Especulação no Mercado**:
   - Investidores financeiros aumentaram suas participações em contratos futuros de petróleo, contribuindo para a elevação dos preços.

3. **Crise Financeira Global**:
   - A crise financeira de 2008, desencadeada pelo colapso do Lehman Brothers em setembro, resultou em uma recessão global que reduziu drasticamente a demanda por petróleo, levando à queda acentuada dos preços.

4. **Flutuações Cambiais**:
   - A desvalorização do dólar americano no início de 2008 tornou o petróleo mais barato para detentores de outras moedas, estimulando a demanda e elevando os preços.

### **Análise da Queda do Preço do Petróleo em 2014**

#### **Contexto Histórico**
Em 2014, o mercado global de petróleo registrou uma queda dramática nos preços, com o barril do Brent passando de aproximadamente **US 115 em junho** para cerca de **US 46 em janeiro de 2015**. Essa queda foi causada por uma combinação de fatores estruturais e estratégicos que resultaram em um excesso de oferta no mercado global.

#### **Impacto no Mercado de Petróleo**
1. **Queda Acentuada nos Preços**:
   - Em seis meses, o preço do petróleo Brent caiu mais de **60%**, marcando uma das quedas mais rápidas e significativas da história recente.

2. **Desequilíbrio entre Oferta e Demanda**:
   - O aumento da produção global coincidiu com uma desaceleração na demanda por petróleo, especialmente em economias emergentes.

#### **Fatores Contribuintes**
1. **Produção Aumentada de Petróleo de Xisto nos Estados Unidos**:
   - Avanços em tecnologias como o fraturamento hidráulico (fracking) permitiram aos Estados Unidos aumentar significativamente a produção de petróleo de xisto.
   - O aumento da produção americana reduziu a dependência de importações e adicionou uma grande quantidade de petróleo ao mercado global, criando um excedente de oferta.

2. **Decisão Estratégica da OPEP**:
   - Em novembro de 2014, a Organização dos Países Exportadores de Petróleo (OPEP), liderada pela Arábia Saudita, decidiu manter os níveis de produção, apesar do excesso de oferta no mercado.
   - Essa decisão foi vista como uma tentativa de pressionar produtores de alto custo, como os de petróleo de xisto, a sair do mercado, preservando a participação da OPEP a longo prazo.

3. **Redução da Demanda Global**:
   - A desaceleração econômica em economias emergentes, particularmente na China, resultou em uma demanda menor por petróleo, agravando o desequilíbrio no mercado.

4. **Fortalecimento do Dólar Americano**:
   - O fortalecimento do dólar americano tornou o petróleo, cotado nessa moeda, mais caro para compradores estrangeiros, reduzindo a demanda.

#### **Consequências**
1. **Impacto nos Países Produtores**:
   - Economias altamente dependentes da exportação de petróleo, como Venezuela, Nigéria e Rússia, sofreram grandes perdas de receita, enfrentando desafios econômicos severos.
   - Esses países foram forçados a adotar medidas de austeridade ou buscar alternativas econômicas.

2. **Benefícios para Países Importadores**:
   - Países importadores de petróleo, como Estados Unidos e membros da União Europeia, se beneficiaram da queda nos preços, resultando em custos de energia mais baixos, maior consumo e redução da inflação.

### **Análise da Queda do Preço do Petróleo em 2020**

#### **Contexto Histórico**
Em 2020, o mercado global de petróleo enfrentou uma das maiores quedas de preços já registradas, resultando de uma combinação de fatores sem precedentes que afetaram tanto a oferta quanto a demanda.

#### **Impacto no Mercado de Petróleo**
1. **Queda Acentuada nos Preços**:
   - Em abril de 2020, o preço do barril de petróleo Brent caiu para menos de **US$ 10**, níveis não vistos desde 2002.

2. **Preços Negativos do WTI**:
   - Em 20 de abril de 2020, os contratos futuros do petróleo WTI nos EUA atingiram valores negativos pela primeira vez na história, chegando a **-US$ 37,63** por barril.

#### **Fatores Contribuintes**
1. **Pandemia de COVID-19**:
   - **Redução da Demanda**: Medidas de confinamento e restrições de viagem globalmente implementadas reduziram drasticamente a demanda por combustíveis fósseis.
   - **Interrupção Econômica**: A desaceleração econômica global diminuiu a atividade industrial, reduzindo ainda mais a demanda por energia.

2. **Guerra de Preços entre Arábia Saudita e Rússia**:
   - **Colapso das Negociações da OPEP+**: Em março de 2020, a OPEP e aliados, incluindo a Rússia, não conseguiram chegar a um acordo sobre cortes de produção.
   - **Aumento da Produção**: A Arábia Saudita respondeu aumentando sua produção e oferecendo descontos significativos, iniciando uma guerra de preços que exacerbou a queda dos preços.

3. **Capacidade de Armazenamento Limitada**:
   - Com a demanda em declínio e a produção permanecendo alta, os estoques de petróleo atingiram a capacidade máxima, especialmente nos EUA, levando a situações onde os produtores pagavam para que outros assumissem o petróleo excedente.

#### **Consequências**
1. **Impacto nos Países Produtores**:
   - Países dependentes da exportação de petróleo, como Rússia, Venezuela e países do Oriente Médio, enfrentaram quedas significativas nas receitas, afetando suas economias.

2. **Benefícios para Países Importadores**:
   - Nações importadoras de petróleo experimentaram custos de energia mais baixos, o que poderia estimular a recuperação econômica pós-pandemia.

3. **Impacto na Indústria de Energia**:

```python
dfp = dfb.sort_index()
dfp = dfp.reset_index()
dfp
```

```python
# Aplicando o modelo Prophet
model_prophet = Prophet()
model_prophet.fit(dfp)
future = model_prophet.make_future_dataframe(periods=90, freq='B') # Vamos usar um periodo de 90 dias para frente, com uma frequencia B de dias uteis
forecast = model_prophet.predict(future) # Criando as previsoes.
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error # Importando a lib para termos os parametros de como foi o forecast
```

```python
model_prophet.plot(forecast)
```

```python
model_prophet.plot_components(forecast)
```

Conforme identificado na Análise Exploratória de Dados (EDA), a aplicação do modelo Prophet corrobora os insights previamente observados na análise mensal: os meses de dezembro e janeiro tendem a apresentar valores mais baixos, alinhando-se às explicações já discutidas.

Adicionalmente, o modelo proporciona uma nova perspectiva sobre o comportamento semanal dos preços. Observa-se uma tendência inicial de queda desde o início da semana, com os valores médios atingindo seu ponto mais baixo às quartas-feiras. A partir desse ponto, os preços passam a se recuperar gradualmente, alcançando os valores mais altos no final da semana. Essa dinâmica reflete o comportamento típico do mercado, considerando a ausência de cotações durante os finais de semana.

```python
import numpy as np
# Divisão dos dados em treino e teste
train = dfp.iloc[:-30]  # Treino: Dados até os últimos 30 pontos
test = dfp.iloc[-30:]   # Teste: Últimos 30 pontos

# Ajustando o modelo Prophet com o conjunto de treino
model_normal = Prophet()
model_normal.fit(train)

# Criando datas futuras para prever o conjunto de teste
future = model_normal.make_future_dataframe(periods=30, freq='D')  # Previsão para o período de teste
forecast = model_normal.predict(future)

# Filtrar apenas as previsões correspondentes ao conjunto de teste
forecast_test = forecast.iloc[-30:]

# Extraindo as previsões e valores reais
y_true = test['y'].values  # Valores reais no conjunto de teste
y_pred = forecast_test['yhat'].values  # Valores previstos no mesmo período

# Calculando métricas
prophet_mae = mean_absolute_error(y_true, y_pred)
prophet_rmse = np.sqrt(mean_squared_error(y_true, y_pred))
prophet_mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
print(f"Prophet MAE: {prophet_mae:.2f}, RMSE: {prophet_rmse:.2f}, MAPE: {prophet_mape:.2f}%")
```

### **3. MAPE (Mean Absolute Percentage Error)**: **21,16%**
